```python
ndf=pd.DataFrame()
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_values = scaler.fit_transform(pdf[['MDVP:Fo(Hz)', 'MDVP:Fhi(Hz)', 'MDVP:Flo(Hz)', 'MDVP:Jitter(%)',
       'MDVP:Jitter(Abs)', 'MDVP:RAP', 'MDVP:PPQ', 'Jitter:DDP',
       'MDVP:Shimmer', 'MDVP:Shimmer(dB)', 'Shimmer:APQ3', 'Shimmer:APQ5',
       'MDVP:APQ', 'Shimmer:DDA', 'NHR', 'HNR', 'RPDE', 'DFA',
       'spread1', 'spread2', 'D2', 'PPE']])
c=0
for i in list(pdf.columns):
    if i!='status':
        ndf[i] = scaled_values[:,c]
        c=c+1
    else:
        ndf[i]=pdf[i]
    
```

```python
from sklearn.model_selection import train_test_split
ndf_train,ndf_test=train_test_split(ndf,test_size=0.2,random_state=100)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
features=selected2
tar=['status']
model.fit(ndf_train[features],ndf_train[tar])
```

```python
train_accuracy = model.score(ndf_train[features],ndf_train[tar])
test_accuracy = model.score(ndf_test[features],ndf_test[tar])
print(train_accuracy,test_accuracy)
```
